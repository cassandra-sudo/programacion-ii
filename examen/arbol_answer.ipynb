### La clase `Nodo`

La clase `Nodo` y `AB` (Árbol Binario) son definidas, y con base a estas clases, es necesario implementar las clases `ABB` (Árbol Binario de Búsqueda) y `AAVL` (Árbol Balanceado). Haciendo uso de los conceptos de herencia y programación orientada a objetos (POO), es necesario definir la clase `ABB` y sus respectivas operaciones, de igual manera es necesario definir la clase `AVVL` heredando de la clase `ABB`.

**ESTRICTAMENTE PROHIBIDO HACER USO DE FUNCIONES QUE NO HAYAS DEFINIDO**.

A continuación se define la clase `Nodo` que cumple con la finalidad de definir la estructura básica para crear estructuras más elaboradas como son los Árboles Binarios de Búsqueda o los Árboles Balanceados.

```python
# Clase Nodo necesaria para generar arboles binarios
class Nodo(object):
  '''
  Constructor de nodos
  dato: es el contenido del nodo
  '''
  def __init__(self, dato):
    self.izq = None
    self.dato = dato
    self.der = None
    self.padre = None

  # Nos dice si el nodo tiene al menos un hijo    
  def tieneHijos(self):
    return self.izq is not None and self.der is not None
  
  # Nos dide si el nodo es un hijo izquierdo
  def esIzq(self):
    return self.padre.izq == self
  
  # Nos dice si el nodo es un hijo derecho
  def esDer(self):
    return self.padre.der == self
  
  # Imprime el contenido del nodo
  def __str__(self):
    if self.dato is None:
      pass
    else:
      return '{}'.format(self.dato)
```

### La clase `AB` Árboles Binarios

Se define la clase `AB` la cual es una implementación de la estructura de datos conocida como Árbol Binario.

Esta clase ya cuenta con algunos atributos (raíz) y algunos métodos que se van a utilizar en las siguientes clases.

```python
import random

''' Clase AB (ArbolBinario), genera la estructura de un arbol binario, asi como sus funciones.'''
class AB(object):

  ''' Consructor de arboles binarios'''
  def __init__(self, dato):
    self.raiz = Nodo(dato)
      
  ''' Elimina todo el arbol'''
  def elimina(self):
    self.raiz = None
      
  ''' Devuelve el nivel del nodo que se le pasa como parametro.
      La raiz se ubica en el nivel cero'''
  def nivel(self, nodo):
    if nodo is None:
      return -1
    else:
      return 1 + self.nivel(nodo.padre)

  ''' Devuelve la altura a partir del nodo que se le pasa como parametro,
      si el arbol es vacio la altura es cero, si no se le suma 1 al maximo
      de las alturas de sus hijos'''
  def altura(self, nodo):
    if nodo is None:
      return 0
    else:
      return 1 + max(self.altura(nodo.izq), self.altura(nodo.der))
      
  ''' Inserta recursivamente un dato en el arbol de manera recursiva
      nodo: es el nodo a partir del cual se quiere insertar
      dato: es el dato que se quiere insertar
  '''
  def insertar_nodo(self, nodo, dato):
    # si el nodo es vacio ahi se crea el nuevo nodo
    if nodo is None:
      nuevo_nodo = Nodo(dato)
      return nuevo_nodo
    # dado que en los AB no hay un orden el nuevo nodo se inserta donde sea
    if bool(random.getrandbits(1)):
      nuevo_nodo = self.insertar_nodo(nodo.izq, dato)
      nodo.izq = nuevo_nodo
      nuevo_nodo.padre = nodo
    else:
      nuevo_nodo = self.insertar_nodo(nodo.der, dato)
      nodo.der = nuevo_nodo
      nuevo_nodo.padre = nodo
    #nodo guarda toda la ruta de donde sera insertado el dato
    #hasta caer en el caso base, es por eso que se devuelve    
    return nodo
  
  ''' Inserta un nodo en el arbol, a partir de la raiz'''
  def insertar(self, dato):
    #Se inserta el dato desde la raiz
    self.insertar_nodo(self.raiz, dato)
      
  ''' Recorre el arbol enorden e imprime cada uno de sus nodos'''
  def recorrido_enorden(self, nodo, recorrido):
    if nodo is not None:
      self.recorrido_enorden(nodo.izq, recorrido)
      recorrido.append(nodo.dato)
      self.recorrido_enorden(nodo.der, recorrido)

  ''' Recorre el arbol preorden e imprime cada uno de sus nodos'''
  def recorrido_preorden(self, nodo, recorrido):
    if nodo is not None:
      recorrido.append(nodo.dato)
      self.recorrido_preorden(nodo.izq, recorrido)
      self.recorrido_preorden(nodo.der, recorrido)

  ''' Recorre el arbol postorden e imprime cada uno de sus nodos'''
  def recorrido_postorden(self, nodo, recorrido):
    if nodo is not None:
      self.recorrido_postorden(nodo.izq, recorrido)
      self.recorrido_postorden(nodo.der, recorrido)
      recorrido.append(nodo.dato)
```

### La clase `ABB` Árboles Binarios de Búsqueda

Haciendo uso de herencia, y tomando como base la clase `AB` define la clase `ABB` la cual es una implementación de la estructura de datos conocida como Árbol Binario de Búsqueda.

La siguiente celda solo tiene la _firma_ de los métodos, es neceario que completes el código para que la clase `ABB` cumpla con todos los requeriminetos y pase la pruebas. Los métodos a implemetar son:

- `busqueda`: recibe un nodo a partir del cual buscar y un dato a buscar y devuelve el nodo en el cual se encuentra el dato o `None` en caso de que no exista el dato.
- `insetar`: Recibe un dato y este se intenta insertar desde la raíz, **no se permiten datos duplicados**.
- `borrar`: Recibe un dato, mismo que se intenta borrar en caso de que no exista.

Puedes apoyarte en la clase `AB` si es que así lo consideras necesario.

```python
''' Clase arbol binario de busqueda, genera la estructura de un arbol binario
    de busqueda asi como sus funciones.'''
class ABB(AB):

  def busqueda(self, nodo, dato):
    '''
    Busca un dato en el arbol a partir del nodo dado.
    Retorna el nodo que contiene el dato, o None si no existe.
    '''
    # Caso base: nodo vacio -> dato no encontrado
    if nodo is None:
      return None
    # Caso base: encontramos el dato
    if dato == nodo.dato:
      return nodo
    # Si el dato es menor, buscar en el subarbol izquierdo
    elif dato < nodo.dato:
      return self.busqueda(nodo.izq, dato)
    # Si el dato es mayor, buscar en el subarbol derecho
    else:
      return self.busqueda(nodo.der, dato)

  def insertar(self, dato):
    '''
    Inserta un dato en el ABB respetando la propiedad de orden.
    No se permiten datos duplicados.
    '''
    self.raiz = self._insertar_abb(self.raiz, dato, None)

  def _insertar_abb(self, nodo, dato, padre):
    '''
    Funcion auxiliar recursiva para insertar en el ABB.
    Retorna el nodo raiz del subarbol modificado.
    '''
    # Caso base: posicion encontrada, crear nuevo nodo
    if nodo is None:
      nuevo = Nodo(dato)
      nuevo.padre = padre
      return nuevo
    # Dato duplicado: no se inserta
    if dato == nodo.dato:
      return nodo
    # Insertar en subarbol izquierdo
    elif dato < nodo.dato:
      nodo.izq = self._insertar_abb(nodo.izq, dato, nodo)
    # Insertar en subarbol derecho
    else:
      nodo.der = self._insertar_abb(nodo.der, dato, nodo)
    return nodo

  def _minimo(self, nodo):
    '''
    Retorna el nodo con el valor minimo del subarbol dado.
    En un ABB el minimo siempre esta en el nodo mas a la izquierda.
    '''
    actual = nodo
    while actual.izq is not None:
      actual = actual.izq
    return actual

  def borrar(self, dato):
    '''
    Elimina el nodo con el dato dado del ABB.
    Si el dato no existe, el arbol no se modifica.
    '''
    self.raiz = self._borrar_abb(self.raiz, dato)

  def _borrar_abb(self, nodo, dato):
    '''
    Funcion auxiliar recursiva para borrar un nodo del ABB.
    Maneja los tres casos: nodo hoja, nodo con un hijo, nodo con dos hijos.
    Retorna la raiz del subarbol modificado.
    '''
    # Caso base: el dato no existe
    if nodo is None:
      return None

    if dato < nodo.dato:
      # Buscar en subarbol izquierdo
      nodo.izq = self._borrar_abb(nodo.izq, dato)
      if nodo.izq is not None:
        nodo.izq.padre = nodo
    elif dato > nodo.dato:
      # Buscar en subarbol derecho
      nodo.der = self._borrar_abb(nodo.der, dato)
      if nodo.der is not None:
        nodo.der.padre = nodo
    else:
      # Nodo encontrado: tres casos posibles

      # Caso 1: nodo hoja (sin hijos)
      if nodo.izq is None and nodo.der is None:
        return None

      # Caso 2a: solo tiene hijo derecho
      elif nodo.izq is None:
        nodo.der.padre = nodo.padre
        return nodo.der

      # Caso 2b: solo tiene hijo izquierdo
      elif nodo.der is None:
        nodo.izq.padre = nodo.padre
        return nodo.izq

      # Caso 3: tiene dos hijos
      # Se reemplaza el dato con el sucesor inorden (minimo del subarbol derecho)
      else:
        sucesor = self._minimo(nodo.der)
        nodo.dato = sucesor.dato
        # Borrar el sucesor del subarbol derecho
        nodo.der = self._borrar_abb(nodo.der, sucesor.dato)
        if nodo.der is not None:
          nodo.der.padre = nodo

    return nodo
```

#### `busqueda`

¿Cuál es el peor caso cuando se busca un dato en un árbol binario de búsqueda y a qué orden de complejidad respecto al tiempo pertenece el método `busqueda`. ¡Justifica tu respuesta!

> **Respuesta:**
>
> El **peor caso** ocurre cuando el árbol está completamente degenerado, es decir, cuando todos los nodos se insertan en orden creciente o decreciente y el árbol se convierte en una lista enlazada (cada nodo tiene solo un hijo). En ese escenario, buscar un dato requiere recorrer todos los nodos desde la raíz hasta la hoja más profunda.
>
> **Complejidad temporal: O(n)**
>
> Justificación: en el peor caso hay que visitar los *n* nodos del árbol uno por uno. En el caso promedio y mejor caso (árbol balanceado), la búsqueda descarta la mitad del árbol en cada paso, dando una complejidad de **O(log n)**. Sin embargo, el peor caso teórico sigue siendo O(n) para un ABB sin balanceo.

#### `insertar`

¿A qué orden de complejidad respecto al tiempo y a la memoria pertenece el método `insertar`. ¡Justifica tu respuesta!

> **Respuesta:**
>
> **Complejidad temporal: O(n)** en el peor caso, **O(log n)** en el caso promedio.
>
> Justificación: insertar un dato requiere primero encontrar la posición correcta recorriendo el árbol desde la raíz. En un árbol balanceado esto toma O(log n) comparaciones, pero en un árbol degenerado (lista enlazada) puede requerir recorrer los *n* nodos. El trabajo en cada nodo visitado es constante O(1).
>
> **Complejidad en memoria: O(n)** en el peor caso, **O(log n)** en el caso promedio.
>
> Justificación: la implementación es recursiva, por lo que se apila un frame de llamada por cada nivel recorrido. En el peor caso (árbol degenerado) la pila de recursión alcanza profundidad *n*; en un árbol balanceado la profundidad es O(log n).

#### `borrar`

¿A qué orden de complejidad respecto al tiempo y a la memoria pertenece el método `borrar`. ¡Justifica tu respuesta!

> **Respuesta:**
>
> **Complejidad temporal: O(n)** en el peor caso, **O(log n)** en el caso promedio.
>
> Justificación: borrar un nodo implica primero buscarlo (misma complejidad que `busqueda`) y luego realizar la reestructuración del árbol. En el caso de dos hijos, se necesita encontrar el sucesor inorden (mínimo del subárbol derecho), lo cual es también O(h) donde h es la altura. En total, toda la operación es proporcional a la altura del árbol: O(log n) promedio, O(n) peor caso.
>
> **Complejidad en memoria: O(n)** en el peor caso, **O(log n)** en el caso promedio.
>
> Justificación: igual que en `insertar`, la recursión consume espacio en la pila de llamadas proporcional a la altura del árbol.

#### `elimina`

¿A qué orden de complejidad respecto al tiempo y a la memoria pertenece el método `elimina`. ¡Justifica tu respuesta!

> **Respuesta:**
>
> **Complejidad temporal: O(1)**
>
> Justificación: el método `elimina` de la clase `AB` únicamente asigna `None` a `self.raiz`. Esta es una operación de asignación directa que no depende del tamaño del árbol, por lo que es de tiempo constante. El recolector de basura de Python se encargará de liberar los nodos en algún momento posterior, pero eso es transparente al programa.
>
> **Complejidad en memoria: O(1)**
>
> Justificación: la operación no usa espacio adicional proporcional al tamaño del árbol; solo reasigna un atributo.

### La clase `Árbol AVL` Árboles Balanceados

Heredando de la clase `ABB` es necesario definir la clase `ArbolAVL` misma que es una implementación de la estructura de datos conocida como Arboles Balanceados.

Dado que la forma en la que se insertan y eliminan datos en un Arbol Balanceado es diferente a como se lleva a cabo en los Arboles Binarios de Búsqueda, es necesario redefinir algunos métodos. Los métodos a implemetar son:

- `insertar`: Una vez que se inserta un dato, es necesario rebalancear el árbol.
- `borrar`: Una vez que se borra un dato, es necesario rebalancear el árbol.

```python
class ArbolAVL(ABB):

  # ------------------------------------------------------------------ #
  # Utilidades internas de balanceo                                      #
  # ------------------------------------------------------------------ #

  def _altura_nodo(self, nodo):
    '''Retorna la altura del nodo dado; 0 si es None.'''
    return self.altura(nodo)

  def _factor_balance(self, nodo):
    '''
    Calcula el factor de balance de un nodo:
    altura(subarbol izquierdo) - altura(subarbol derecho).
    Un arbol AVL mantiene este valor en {-1, 0, 1} para cada nodo.
    '''
    if nodo is None:
      return 0
    return self._altura_nodo(nodo.izq) - self._altura_nodo(nodo.der)

  def _rotar_derecha(self, z):
    '''
    Rotacion simple a la derecha alrededor del nodo z.
    Se usa cuando el subarbol izquierdo esta mas alto (desbalance a la izquierda).

            z                y
           / \              / \
          y   T4    -->    x   z
         / \                  / \
        x   T3              T3  T4
    '''
    y = z.izq
    T3 = y.der

    # Realizar rotacion
    y.der = z
    z.izq = T3

    # Actualizar padres
    y.padre = z.padre
    z.padre = y
    if T3 is not None:
      T3.padre = z

    return y

  def _rotar_izquierda(self, z):
    '''
    Rotacion simple a la izquierda alrededor del nodo z.
    Se usa cuando el subarbol derecho esta mas alto (desbalance a la derecha).

        z                  y
       / \                / \
      T1   y    -->      z   x
          / \           / \
        T2   x        T1  T2
    '''
    y = z.der
    T2 = y.izq

    # Realizar rotacion
    y.izq = z
    z.der = T2

    # Actualizar padres
    y.padre = z.padre
    z.padre = y
    if T2 is not None:
      T2.padre = z

    return y

  def _rebalancear(self, nodo):
    '''
    Verifica y corrige el balance del nodo aplicando la rotacion necesaria.
    Retorna la nueva raiz del subarbol (puede cambiar tras la rotacion).
    Cubre los cuatro casos AVL: LL, RR, LR, RL.
    '''
    fb = self._factor_balance(nodo)

    # Caso LL: desbalance izquierda-izquierda -> rotacion simple derecha
    if fb > 1 and self._factor_balance(nodo.izq) >= 0:
      return self._rotar_derecha(nodo)

    # Caso LR: desbalance izquierda-derecha -> rotar hijo izquierdo a la izq, luego raiz a la der
    if fb > 1 and self._factor_balance(nodo.izq) < 0:
      nodo.izq = self._rotar_izquierda(nodo.izq)
      if nodo.izq is not None:
        nodo.izq.padre = nodo
      return self._rotar_derecha(nodo)

    # Caso RR: desbalance derecha-derecha -> rotacion simple izquierda
    if fb < -1 and self._factor_balance(nodo.der) <= 0:
      return self._rotar_izquierda(nodo)

    # Caso RL: desbalance derecha-izquierda -> rotar hijo derecho a la der, luego raiz a la izq
    if fb < -1 and self._factor_balance(nodo.der) > 0:
      nodo.der = self._rotar_derecha(nodo.der)
      if nodo.der is not None:
        nodo.der.padre = nodo
      return self._rotar_izquierda(nodo)

    # Nodo ya esta balanceado
    return nodo

  # ------------------------------------------------------------------ #
  # insertar                                                             #
  # ------------------------------------------------------------------ #

  def insertar(self, dato):
    '''
    Inserta un dato en el arbol AVL y rebalancea si es necesario.
    No se permiten datos duplicados (heredado del ABB).
    '''
    self.raiz = self._insertar_avl(self.raiz, dato, None)
    # Asegurar que la raiz no tenga padre
    if self.raiz is not None:
      self.raiz.padre = None

  def _insertar_avl(self, nodo, dato, padre):
    '''
    Funcion auxiliar recursiva: inserta y rebalancea hacia arriba.
    Retorna la raiz del subarbol modificado y balanceado.
    '''
    # Caso base: posicion vacia
    if nodo is None:
      nuevo = Nodo(dato)
      nuevo.padre = padre
      return nuevo

    # Insercion BST estandar
    if dato < nodo.dato:
      nodo.izq = self._insertar_avl(nodo.izq, dato, nodo)
      if nodo.izq is not None:
        nodo.izq.padre = nodo
    elif dato > nodo.dato:
      nodo.der = self._insertar_avl(nodo.der, dato, nodo)
      if nodo.der is not None:
        nodo.der.padre = nodo
    else:
      # Dato duplicado: no se inserta
      return nodo

    # Rebalancear al regresar en la recursion
    nodo = self._rebalancear(nodo)
    return nodo

  # ------------------------------------------------------------------ #
  # borrar                                                               #
  # ------------------------------------------------------------------ #

  def borrar(self, dato):
    '''
    Elimina un dato del arbol AVL y rebalancea si es necesario.
    Si el dato no existe, el arbol no se modifica.
    '''
    self.raiz = self._borrar_avl(self.raiz, dato)
    if self.raiz is not None:
      self.raiz.padre = None

  def _borrar_avl(self, nodo, dato):
    '''
    Funcion auxiliar recursiva: borra y rebalancea hacia arriba.
    Retorna la raiz del subarbol modificado y balanceado.
    '''
    # Caso base: el dato no existe
    if nodo is None:
      return None

    if dato < nodo.dato:
      nodo.izq = self._borrar_avl(nodo.izq, dato)
      if nodo.izq is not None:
        nodo.izq.padre = nodo
    elif dato > nodo.dato:
      nodo.der = self._borrar_avl(nodo.der, dato)
      if nodo.der is not None:
        nodo.der.padre = nodo
    else:
      # Nodo encontrado

      # Caso 1: nodo hoja
      if nodo.izq is None and nodo.der is None:
        return None

      # Caso 2a: solo hijo derecho
      elif nodo.izq is None:
        nodo.der.padre = nodo.padre
        return nodo.der

      # Caso 2b: solo hijo izquierdo
      elif nodo.der is None:
        nodo.izq.padre = nodo.padre
        return nodo.izq

      # Caso 3: dos hijos -> reemplazar con sucesor inorden
      else:
        sucesor = self._minimo(nodo.der)
        nodo.dato = sucesor.dato
        nodo.der = self._borrar_avl(nodo.der, sucesor.dato)
        if nodo.der is not None:
          nodo.der.padre = nodo

    # Rebalancear al regresar en la recursion
    nodo = self._rebalancear(nodo)
    return nodo
```

#### `busqueda`

¿Cuál es el peor caso cuando se búsca un dato en un árbol binario de búsqueda y a qué orden de complejidad respecto al tiempo pertenece el método `busqueda`. ¡Justifica tu respuesta!

> **Respuesta:**
>
> En un árbol AVL **no existe el peor caso degenerado** del ABB, porque las rotaciones garantizan que el árbol siempre esté balanceado.
>
> **Complejidad temporal: O(log n)** — tanto en el peor caso como en el promedio.
>
> Justificación: la propiedad de balanceo AVL asegura que la altura del árbol es siempre **O(log n)** para *n* nodos. En el peor caso, la búsqueda recorre un camino desde la raíz hasta una hoja, lo cual implica a lo más O(log n) comparaciones. Esto es una mejora garantizada respecto al ABB, donde el peor caso era O(n).
